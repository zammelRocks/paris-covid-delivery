```python
import pandas as pd
import json
import requests
import re
import matplotlib as plt
```

```python

response_API = requests.get('https://opendata.paris.fr/api/records/1.0/search/?dataset=coronavirus-commercants-parisiens-livraison-a-domicile&q=&facet=code_postal&facet=type_de_commerce&facet=fabrique_a_paris&facet=services&rows=2503')

data_fromAPI = response_API.json()

print(type(data_fromAPI))

#print(data_fromAPI)
```

```python
data_fromAPI["parameters"]
#data base parameters
```

```python
data_api = data_fromAPI["records"]
data_api

```

```python
listOfFields= []

for input in data_api:
    listOfFields.append(input["fields"])
listOfFields
```

```python
len(listOfFields)
```

```python
df = pd.DataFrame(listOfFields)
```

```python
#df.tail()
print(df.info())
```

```python
null_Loaded_Data = df.isnull()
null_Loaded_Data
```

```python
#eliminating null values

for col in df.columns:
    df[col].fillna(f"No {col} Provided", inplace = True)
df

```

```python
print(df.dtypes)
```

```python
#cleaning values : Column mail
#library for regex expressions
# looking for valid mails based on their regex

pattern = re.compile(r'([A-Za-z0-9]+[.-_])*[A-Za-z0-9]+@[A-Za-z0-9-]+(\.[A-Z|a-z]{2,})+')

i=0

sumMail=0

#use : re.fullmatch(regex, email) to match mail with formate

for col in df['mail']:
    state = bool(re.fullmatch(pattern, col))
    i=i+1
    ##if col value matches regex defined it's assigned to true 
    #print(col," ***** the value state is ****** ", state)
    if state:
        #print(f"=> row number {i} : " , col )
        sumMail=sumMail+1
    
    
  
print("number of businesses having mails out of 2503 are => ",sumMail)
```

```python
#cleaning values : Column Links
#visualizing with x = re.search("^http.*$", coltype) , link can use https or http
sumLinks=0
links=0

for coltype in df['site_internet']:
    
    links=links+1
    #index 
    #verify if string begins with https
    x = re.search("^http.*$", coltype)
    if x:
        #print(f"=> row number {links} : ",coltype)
        sumLinks=sumLinks+1
        
print("number of entreprises having websites =>", sumLinks)
```

```python
#cleaning values : Column Services
sumLivraison=0
liv=0
for colservices in df['services']:
    liv=liv+1
    #verify if string begins Livraisons à domicile
    x = re.search("^Livraisons à domicile.*$", colservices)
    if x:
        #print(f"=> row number {links} : ",colservices)
        sumLivraison=sumLivraison+1
print("number of entreprises proceeding with Home delivery =>", sumLinks)
```

```python
#cross Table allows us to view the frequency of services used by each type of commerce
crossTab = pd.crosstab(df['type_de_commerce'], df['services'])
crossTab
```

```python
#This data frame can be presented by 
pd.crosstab(df['type_de_commerce'], df['services']).plot(kind='bar')
plt.title('repartition des services par\n type de commerce',
          fontweight ="bold")
```

```python
sum_Type_commerce = crossTab.sum()
aggregation=sum_Type_commerce.to_frame()
aggregation

#Sum  types commerce using this services : aggregation by services
```

```python
#group businesses by domain/Activity
nbr_business_per_domain=df.groupby(['type_de_commerce'])['nom_du_commerce'].count()

X=nbr_business_per_domain.to_frame()
X
print("le domaine ayant max de commerçants", X['nom_du_commerce'].max())

X['nom_du_commerce']= X['nom_du_commerce']*100/2503

X.head()
```

```python
index_list=[]
i=-1
for idx in X.index:
    i=i+1
    index_list.append(idx)

#print(index_list)
    
pourcentage_list= X['nom_du_commerce'].to_list()

data = {'type_de_commerce':index_list,'shop_pourcentage':pourcentage_list}
newdf= pd.DataFrame(data)

newdf.head()
```

```python
plt.figure(figsize=(10,5))
hist=plt.bar(newdf['type_de_commerce'], newdf['shop_pourcentage'],0.8,color='red')
plt.xlabel('type_de_commerce')
plt.ylabel('shop_pourcentage')
plt.xticks(rotation=90)
plt.title('Pourcentage de commerçant \n par secteur de travail',
          fontweight ="bold")
hist
```
